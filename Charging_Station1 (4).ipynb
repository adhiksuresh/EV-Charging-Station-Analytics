```python

# Convert 'reading_time' column to datetime format
filtered_march['Reading time'] = pd.to_datetime(filtered_march['Reading time'])

# Define tariff rates
tariff_rates = {
    'off-peak': 6,
    'peak': 8,
    'mid-peak': 7,
    'high-peak': 8
}

# Function to determine the tariff based on the time of day
def get_tariff(reading_time):
    hour = reading_time.hour
    if 23 <= hour or hour < 6:
        return 'off-peak'
    elif 7 <= hour < 11:
        return 'peak'
    elif 11 <= hour < 19:
        return 'mid-peak'
    else:
        return 'high-peak'

# Apply the get_tariff function to each row to get the tariff for each reading
filtered_march['tariff'] = filtered_march['Reading time'].apply(get_tariff)

# Calculate the price for each 15-minute interval
filtered_march['price'] = filtered_march.apply(lambda row: row['Difference'] * tariff_rates[row['tariff']], axis=1)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Assuming your DataFrame is called 'filtered_march'
# Make sure 'Reading time' column is in datetime format
filtered_march['Reading time'] = pd.to_datetime(filtered_march['Reading time'])

# Extract relevant columns and set 'Reading time' as the index
df = filtered_march[['Reading time', 'Difference']].set_index('Reading time')

# Resample the data to hourly frequency and sum the energy consumption for each hour
hourly_energy = df.resample('H').sum()

# Normalize the data
scaler = MinMaxScaler()
scaled_energy = scaler.fit_transform(hourly_energy)

# Function to create time series dataset
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)

# Define the look-back period (number of previous hours to consider)
look_back = 24  # For example, using the past 24 hours to predict the next hour

# Create the dataset with the look-back period
X, Y = create_dataset(scaled_energy, look_back)

# Reshape the input data for LSTM
X = np.reshape(X, (X.shape[0], 1, X.shape[1]))

# Create and compile the LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(1, look_back)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X, Y, epochs=50, batch_size=1, verbose=2)

# Forecast energy consumption for a single day
# Use the last 24 hours of available data to predict the next hour
last_24_hours = scaled_energy[-look_back:]
forecast_input = np.reshape(last_24_hours, (1, 1, look_back))
forecast = model.predict(forecast_input)

# Invert the forecast to get it in the original scale
forecast = scaler.inverse_transform(np.reshape(forecast, (forecast.shape[0], 1)))

print(f"Forecasted energy consumption for the next hour: {forecast[0][0]}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with columns: 'reading_time', 'Difference', 'Power(kW)'
# You may need to load your actual dataset using pd.read_csv or another appropriate method.
filtered_march['Reading time'] = pd.to_datetime(filtered_march['Reading time'])

# Extract date from the timestamp
filtered_march['date'] = filtered_march['Reading time'].dt.date

# Calculate average and peak load for each day
daily_load = filtered_march.groupby('date')['Power'].agg(['mean', 'max'])

# Calculate load factor for each day
daily_load['load_factor'] = (daily_load['mean'] / daily_load['max']) * 100
print("Load Factor for Each Day:")
print(daily_load[['load_factor']])

# Plot the graph with improved aesthetics
plt.figure(figsize=(10, 6))
plt.plot(daily_load.index, daily_load['load_factor'], marker='o', linestyle='-', color='b', label='Load Factor')
plt.title('Daily Load Factor')
plt.xlabel('Date')
plt.ylabel('Load Factor (%)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
day_labels = ['Day {}'.format(i + 1) for i in range(len(daily_load))]
plt.xticks(daily_load.index, day_labels, rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Convert 'reading_time' column to datetime format
filtered_march['Reading time'] = pd.to_datetime(filtered_march['Reading time'])

# Define tariff rates
tariff_rates = {
    'off-peak': 10,
    'peak': 20,
    'mid-peak': 15,
    'high-peak': 25
}

# Function to determine the tariff based on the time of day
def get_tariff(reading_time):
    hour = reading_time.hour
    if 23 <= hour or hour < 6:
        return 'off-peak'
    elif 7 <= hour < 11:
        return 'peak'
    elif 11 <= hour < 19:
        return 'mid-peak'
    else:
        return 'high-peak'

# Apply the get_tariff function to each row to get the tariff for each reading
filtered_march['tariff'] = filtered_march['Reading time'].apply(get_tariff)

# Calculate the price for each 15-minute interval
filtered_march['evprice'] = filtered_march.apply(lambda row: row['Difference'] * tariff_rates[row['tariff']], axis=1)
```

```python
# Calculate the total of the 'price' column
total_price = filtered_march['evprice'].sum()
total_energy = filtered_march['Difference'].sum()
# Print the total
print(f"Total Energy :{round(total_energy,2)} Units")
print("Total Price Rs:", round(total_price,2))
```
